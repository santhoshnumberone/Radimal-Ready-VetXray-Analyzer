### Import Libraries

```python
import pydicom
import matplotlib.pyplot as plt
```

###### Define the path to your DICOM file.

```python
# Ensure this path is correct for your file
dicom_file_path = '../data/image-00001.dcm'
```

###### Read DICOM File

```python
ds = pydicom.dcmread(dicom_file_path)
```

###### Print the entire dataset to display all the metadata tags. Then, print a few key ones individually to show you can access specific information.

```python
print(ds)
print("\n--- Key Metadata Tags ---")
print(f"Patient ID: {ds.PatientID}")
print(f"Study Date: {ds.StudyDate}")
print(f"Modality: {ds.Modality}")
```

###### Extract the pixel data and display it using matplotlib. The pixel_array attribute is what we need.

```python
# Get the image data as a numpy array
image_data = ds.pixel_array

# Display the image
plt.imshow(image_data, cmap='gray')
plt.title(f"DICOM Image (Modality: {ds.Modality})")
plt.show()
```
